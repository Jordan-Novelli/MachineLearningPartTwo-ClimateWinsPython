### 3. Dendogram

#### Using PCA Reduced Data
##### I'll be using the PCA that had 2 variables to create more unique visuals. I'll also be using KASSEL again to keep the data relation to the previous Hierarchiacal Clustering.

#### Single

```python
# Single Dendogram (using dfPCA data)

distance_single = linkage(dfPCA.loc[:,['KASSEL_pleasant_weather']],method='single')
plt.figure(figsize=(18,6))
dendrogram(distance_single, leaf_rotation=90)
plt.xlabel('Index')
plt.ylabel('Distance')
plt.suptitle("Dendrogram Single Method",fontsize=18)
plt.savefig('Dendrogram Single Method - dfPCA Kassel Year 2000.png')
plt.show()
```

```python
dfPCA_SM=dfPCA.copy()

dfPCA_SM['clust2']=fcluster(distance_single,2, criterion='maxclust')
dfPCA_SM['clust3']=fcluster(distance_single,3, criterion='maxclust')
dfPCA_SM.head()
```

```python
# Plot the clusters against the answers

plt.figure(figsize=(25,8))

plt.suptitle("Single Method Results",fontsize=18)
plt.subplot(1,3,3)
plt.title("Kassel 2022",fontsize=14)
sns.scatterplot(x="PCA1",y="PCA2", data=dfPCA_SM, hue="KASSEL_pleasant_weather")
plt.savefig('Scatter Plot Kassel 2000_dfPCA_single.png')
```

```python
plt.figure(figsize=(20,5))
plt.title("K = 2",fontsize=14)
sns.swarmplot(x="PCA1",y="clust2", data=dfPCA_SM, hue="KASSEL_pleasant_weather")
plt.savefig('Swarm Plot Kassel_dfPCA_SM_2cluster_single.png')

plt.figure(figsize=(20,5))
plt.title("K = 3",fontsize=14)
sns.swarmplot(x="PCA1",y="clust3", data=dfPCA_SM, hue="KASSEL_pleasant_weather")
plt.savefig('Swarm Plot KASSEL_dfPCA_SM_3cluster_single.png')
```
